## Imports

```
import numpy as np
import pandas as pd

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical, plot_model
from keras.datasets import fashion_mnist
from sklearn.metrics import classification_report

import matplotlib.pyplot as plt
%matplotlib inline
```

## Trainning and Test Data

```
(train_x, train_y), (test_x_img, test_y) = fashion_mnist.load_data()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# show the first image of trainning data
plt.figure()
plt.imshow(train_x[0])
plt.colorbar()
plt.grid(False)
```

## Reshape the images (matrices) to vectors

```
train_x = train_x.reshape((-1, 784))
test_x = test_x_img.reshape((-1, 784))
```

## MLP Neural Network Models

### 3 Layers model

```
network_3 = Sequential([
                        Dense(28, activation='relu', input_shape=(784,)),
                        Dense(128, activation='relu'),
                        Dense(10, activation='softmax')
])

network_3.compile(
    optimizer='adam', 
    loss='categorical_crossentropy', 
    metrics=['categorical_accuracy']
)

network_3.summary()
```

### 6 Layers model

```
network_6 = Sequential([
                        Dense(28, activation='relu', input_shape=(784,)),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(10, activation='softmax')
])

network_6.compile(
    optimizer='adam', 
    loss='categorical_crossentropy', 
    metrics=['categorical_accuracy']
)

network_6.summary()
```

### 12 Layers model

```
network_12 = Sequential([
                        Dense(28, activation='relu', input_shape=(784,)),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(128, activation='relu'),
                        Dense(10, activation='softmax')
])

network_12.compile(
    optimizer='adam', 
    loss='categorical_crossentropy', 
    metrics=['categorical_accuracy']
)

network_12.summary()
```

## Trainning the NN Model
20 epochs and batch_size = 50

### 3 Layers

```
history_3 = network_3.fit(
    train_x,
    to_categorical(train_y),
    epochs = 20,
    batch_size = 50,
    validation_split = 0.2
)
```

### 6 Layers

```
history_6 = network_6.fit(
    train_x,
    to_categorical(train_y),
    epochs = 20,
    batch_size = 50,
    validation_split = 0.2
)
```

### 12 Layers

```
history_12 = network_12.fit(
    train_x,
    to_categorical(train_y),
    epochs = 20,
    batch_size = 50,
    validation_split = 0.2
)
```

## Evaluating NN with training data



Creating structure for final table

```
model_analysis = {
    'layers': [],
    'accuracy train data': [],
    'loss train data': [],
    'accuracy test data': [],
    'loss test data': []
}
```

### 3 Layers - Accuracy and loss

```
train_loss, train_acc = network_3.evaluate(
    train_x, 
    to_categorical(train_y)
)
print('Erro na base de teste: ', train_loss)
print('Acurcia na base de teste: ', train_acc)

n_layer = model_analysis['layers']
n_layer.append(3)
model_analysis['layers'] = n_layer

value_accuracy = model_analysis['accuracy train data']
value_accuracy.append(round(train_acc, 2))
model_analysis['accuracy train data'] = value_accuracy

value_loss = model_analysis['loss train data']
value_loss.append(round(train_loss, 2))
model_analysis['loss train data'] = value_loss
```

## Classifications and Predictions

### Funes para exibir imgs e predies

```
# plot image in a grid
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  
  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'
  
  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

# plot the value array    
def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array[i], true_label[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    thisplot= plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0,1])
    predicted_label = np.argmax(predictions_array)
    
    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

### 3 Layers

```
prediction_3 = network_3.predict(test_x)
```

```
y_prediction_3 = prediction_3.argmax(axis=1)
print(classification_report(test_y, y_prediction_3))
```

```
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
plt.title("Predictions of the first 15 images, with NN-3")
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, prediction_3, test_y, test_x_img)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, prediction_3, test_y)
```

## Conclusions

EN-US

A 3-layer neural network has performed poorly in many ways. Its accuracy value for training and test data was approximately $0.75$. In addition, the final error value was approximately $0.60$, a very high value. In the section " Classifications and Predictions", it is possible to observe and prove that this model is not optimized.

However, models of neural networks with 6 and 12 layers propagate well. Its final results were very close, with an accuracy of $0.88$ for training and testing data. In addition, the error for the training and test data resulted in $0.31$ and $0.41$, respectively. Also, the precision, recall and f1-score values are $0.96$, $0.94$ and $0.95$ for 6 layers and $0.94$, $0.96$ and $0.95$ for 12 layers, respectively. Consequently, it is possible to notice that the values are almost equal.

Therefore, in order to consume less computational processing, it is recommended to use neural networks with 6 layers in this problem of Fashion MNIST.
